```python
#cum cream o lista de la 1 la 20?
```

```python
#varianta 1

lista = list(range(1,21)) # simplu, cream o lista de la 1 la 20, precizand ca ultim nr 21

print(lista)
```

```python
#varianta 2

lista = [] # lista goala

for i in range(1,21):
    lista.append(i) # folosim variabila.append ca sa adaugam fiecare element (i) din range in lista goala

print(lista)
```

```python
#varianta 3
lista = [i for i in range(1,21)] # o metodat mai scurta si mai eleganta

print(lista)
```

```python
#cum aflam numarul cel mai mare dintr o lista fara max()?
```

```python
#varianta 1
#presupunem ca primul nr e cel mai mare
lista = [5, 12, 7, 8, 20, 1]

maxim = lista[0]
print("Primul element din lista, care presupunem ca este cel mai mare este", maxim)

for numar in lista:
    if numar > maxim:
        maxim = numar
        # ia toate numerele pe rand, in toate variantele, ajungand intr un final la nr 20

print("Dupa ce am facut bucla, nr cel mai mare este", maxim)
```

```python
#varianta 2

lista.sort() # sortam lista

print(lista)

lista[-1] # nr cu index -1 din lista = ultimul nr, astfel dupa ce o sortam crescator, ultimul e cel mai mare
```

```python
#cum facem suma de lista?
```

```python
#varianta 1

sum(lista) #sum() simplu
```

```python
#varianta 2

suma = 0 # cream o variabila egala cu 0

for numar in lista:
    suma += numar # += aduna fiecare numar din lista in variabila "suma" care era goala

print(suma)
```
